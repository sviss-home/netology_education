```python
import pandas as pd
import numpy  as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv') #чтение данных из url

df.columns =['surgery?', 'Age', 'Hospital Num', 'rect.temp', 'pulse', 'r.rate', 't.of.ext', 'periph pulse', 'mucous membranes', 'capillary refill time', 
             'pain', 'peristalsis', 'abdominal distension', 'nasogastric tube', 'nasogastric reflux', 'nasogastric reflux PH', 'rectal examination - feces', 
             'abdomen', 'packed cell volume', 'total protein', 'abdominocentesis appearance', 'abdomcentesis total protein', 'outcome', 'surgical lesion?',
             'type of lesion1', 'type of lesion2', 'type of lesion3', 'cp_data'] #названия для столбцов

df_cutted = df.loc[:, ['surgery?', 'Age', 'rect.temp', 'pulse', 'r.rate', 't.of.ext', 'pain', 'outcome']] #вывод нужных столбцов
df_values_replaced = df_cutted.replace('?', np.NaN) #замена ? на NaN
df_values_converted = df_values_replaced.astype(float) #преобразование значений датафрейма в тип данных float

for column in df_values_converted:
    columnSeriesObj = df_values_converted[column]
    print('Осн. стат. показатели для колонки',column,':', 'Min =', min(columnSeriesObj.values), 'Max =',max(columnSeriesObj.values), 'Размах =', (max(columnSeriesObj.values) - min(columnSeriesObj.values)), 'Среднее арифметическое =', np.nanmean(columnSeriesObj.values)) #вычисляем min, max, размах и среднее арифметическое, игнорируя NaN

print(df_values_converted.isna().sum()) #подсчитаем число NaN в каждом столбце

df_values_converted.fillna(0).head() #заменяем пропуски на нули, чтобы сохранить датафрейм в изначальном размере
```
