```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv') #создаём датафрейм чтением из файла
ratings_grouped = ratings.groupby(by=['movieId']).mean().sort_values(by='movieId', ascending = True) #находим среднее значение для рейтинга с группировкой по movieId

def classified(row): #определяем функцию сравнения значений в строках датафрейма с заданными в функции значениями
    if row.rating >= 4.5: return 'Высокий рейтинг' #условие1 - возврат заданной в функции строки, если условие выполняется 
    elif 2 < row.rating < 4.5: return 'Средний рейтинг' #...
    elif row.rating <= 2: return 'Низкий рейтинг' #...
     
ratings_grouped['classified'] = ratings_grouped.apply(classified, axis=1) #задаём новый столбец датафрейма, в который будут записаны результаты работы функции
ratings_final = ratings_grouped[['rating','classified']] #фильтруем столбцы, оставляем rating и результаты работы функции
ratings_final.head(3000) #вывод шапки на 3000 значений
```

```python
keywords = pd.read_csv('keywords.csv') #создаём датафрейм чтением из файла
geo_data = { #словарь с ключами и значениями

    'Центр': ['москв', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровcк']

}

def classified(keywords): #определяем функцию проверки наличия ключевых слов в строках датафрейма       
    for region, cities in geo_data.items(): # для ключей region, значений(values) в элементах словаря geo_data
        for city in cities: #итерация city по geo_data values 
            if city in keywords: return region #если итерируемое значение обнаруживается в 
            # else: return 'undefined' # в ином случае возвращаем 'undefined'
     
keywords['classified'] = keywords.apply(classified, axis=1) #задаём новый столбец датафрейма, в который будут записаны результаты работы функции
keywords_sorted = keywords.sort_values('shows', ascending = False) #сортируем датафрейм по данным нового столбца
keywords_sorted.head(550) #вывод шапки на 550 рядов
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
data_merged = ratings.merge(movies, on='movieId', how='left') #объединяем данные, полученные из двух источников данных в один датафрейм data_merged

def production_year(data_merged): #определяем функцию сопоставления условий, заданных в функции, данным объединённого датафрейма
    years = range(1950,2011) #задаём диапазон в локальную пременную years
    for i in years: #запускаем цикл по заданному в переменной years диапазону
        if str(i) in data_merged['title']: return i #если значение из переменной с диапазоном совпадает со значением в строке столбца title то возвращаем его
    else: return '1900' #если не совпадает, то возвращаем 1900

data_merged['year'] = data_merged.apply(production_year, axis=1) #задаём новый столбец year, в который записываем результаты работы функции методом apply с работой построчно сверху вниз       
data_merged_united = data_merged.groupby(by=['year']).mean().sort_values(by='rating', ascending = False) #вычисляем средние значения столбца, предварительно сгруппировав по году, затем сортируем по rating по убыванию 
data_merged_united.head(200) #выводим шапку на 200 строк
```

Работа кода позволяет сделать вывод о том, что, в целом, более новые фильмы имеют рейтинг ниже более старых
